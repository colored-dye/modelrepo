```python
import csv
import requests

skipped_files = ["README.md", ".gitattributes"]

def download_size(repo_id, siblings):
    url = f"https://hf-mirror.com/{repo_id}/resolve/main"
    size = 0
    siblings = eval(siblings)
    for sib in siblings:
        if sib.rfilename in skipped_files:
            continue
        link = f"{url}/{sib.rfilename}"
        r = requests.head(link, headers={"Accept-Encoding": "identity"})
        if r.status_code in (401, 403):
            r = requests.head(link, headers={"Authorization": f"Bearer {HF_TOKEN}", "Accept-Encoding": "identity"})
        if 'location' in r.headers:
            loc = r.headers['location']
            r = requests.head(loc, headers={"Accept-Encoding": "identity"})
        cl = int(r.headers['Content-Length'])
        print(link, cl)
        size += cl
    return size


header = ['id', 'author', 'sha', 'last_modified', 'created_at', 'private', 'gated', 'disabled', 'downloads', 'likes', 'library_name', 'tags', 'pipeline_tag', 'mask_token',
          'card_data', 'widget_data', 'model_index', 'config', 'transformers_info', 'siblings', 'spaces', 'safetensors', 'lastModified', 'cardData', 'transformersInfo', '_id', 'modelId']

siblings_idx = header.index("siblings")

with open("test.csv", "r") as fp:
    reader = csv.reader(fp)
    _ = next(reader)
    for row in reader:
        siblings = row[siblings_idx]
        repo_id = row[0]
        size = download_size(repo_id, siblings)
        print(f"{repo_id}: {size/(1024*1024*1024)} GiB")
```

## Merge all metadata

```python
import csv
import os

metadata_dir = "metadata/likes"

header = ['id', 'author', 'sha', 'last_modified', 'created_at', 'private', 'gated', 'disabled', 'downloads', 'likes', 'library_name', 'tags', 'pipeline_tag', 'mask_token',
          'card_data', 'widget_data', 'model_index', 'config', 'transformers_info', 'siblings', 'spaces', 'safetensors', 'lastModified', 'cardData', 'transformersInfo', '_id', 'modelId']

all_data = []

with open("all.csv", "w") as all_fp:
    writer = csv.writer(all_fp, lineterminator='\n')
    writer.writerow(header)

    for file in os.listdir(metadata_dir):
        file = os.path.join(metadata_dir, file)
        with open(file, "r") as fp:
            reader = csv.reader(fp)
            _ = next(reader)
            rows = list(reader)
            writer.writerows(rows)
        all_data.extend(rows)
```

## Split batches

```python
import csv
import os
import shutil


with open("all.csv", "r") as all_fp:
    reader = csv.reader(all_fp, lineterminator='\n')
    next(reader)

    rows = list(reader)

jobs_dir = "jobs/"
if os.path.exists(jobs_dir):
    shutil.rmtree(jobs_dir)
os.makedirs(jobs_dir)
batch_size = 21
n_batches = (len(rows)/batch_size).__ceil__()
for i in range(n_batches):
    with open(os.path.join(jobs_dir, f"{i}.csv"), "w") as all_fp:
        writer = csv.writer(all_fp, lineterminator='\n')
        writer.writerow(header)
        writer.writerows(rows[i*batch_size:(i+1)*batch_size])
```
